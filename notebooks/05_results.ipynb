# 📊 Bellabeat Case Study: Data Analysis Summary

## 🔍 Dataset Overview

We analyzed 11 datasets related to user activity, sleep, heart rate, and weight:

- `daily_activity`
- `heartrate`
- `hourly_calories`
- `hourly_intensities`
- `hourly_steps`
- `minute_calories`
- `minute_intensities`
- `minute_mets`
- `minute_sleep`
- `minute_steps`
- `weight_log`

**Cleaning Highlights**:
- ✅ No missing or duplicate values (except `Fat` in `weight_log`)
- ✅ Data types were standardized
- ✅ Merged relevant datasets to form a cleaned, unified structure

---

## 📈 Key Findings

### 1. Activity vs. Sedentary Behavior
- Users who walked more had **fewer sedentary minutes**.
- A **negative correlation** was observed between `TotalSteps` and `SedentaryMinutes`.
---

### 2. Weekday vs. Weekend Behavior

| Day Type | Avg Steps | Avg Calories | Avg Sedentary Minutes |
|----------|------------|---------------|-------------------------|
| Weekday  | 8,334      | 2,429         | 665                     |
| Weekend  | 6,645      | 2,308         | 674                     |

- **Weekdays**: higher activity, more calories burned, slightly less sedentary time
- **Weekends**: lower overall activity, slightly better sleep engagement

---

### 3. Time of Day Patterns
- **Afternoon** and **Evening** are peak times for activity.
- **Night** remains consistently low in intensity for all users.

---

### 4. Intensity Deciles by Time & Day Type
- Users in higher deciles are more active, particularly on **weekday afternoons** and **evenings**.
- **Weekend activity** is more balanced but slightly lower overall.

Visualizations (line and bar plots) demonstrated these patterns clearly.

---

## 💡 Insights & Recommendations

1. 🕒 **Evenings and afternoons** are prime activity times. Bellabeat can schedule fitness challenges or reminders accordingly.
2. 📆 **Weekday engagement** is higher. Consider campaigns targeted during the week.
3. 💤 **Sleep tracking is underutilized** — only 23 users logged sleep. Bellabeat should promote this feature.
4. 🪞 **Night hours** are least active — a great opportunity to promote mindfulness and light stretching content.
